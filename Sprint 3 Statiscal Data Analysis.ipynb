## Aggregate data per user

[Now, as the data is clean, aggregate data per user per period in order to have just one record per user per period. It should ease the further analysis a lot.]

```python
# Calculate the number of calls made by each user per month. Save the result.
calls_data['month'] = pd.to_datetime(calls_data['call_date']).dt.month
calls_per_month = calls_data.groupby(['user_id', 'month']).size().reset_index(name='number_of_calls')
```

```python
# Calculate the amount of minutes spent by each user per month. Save the result.
calls_data['duration'] = np.ceil(calls_data['duration'])
minutes_per_month = calls_data.groupby(['user_id', 'month'])['duration'].sum().reset_index(name='total_minutes')
```

```python
# Calculate the number of messages sent by each user per month. Save the result.
messages_data['month'] = pd.to_datetime(messages_data['message_date']).dt.month
messages_per_month = messages_data.groupby(['user_id', 'month']).size().reset_index(name='number_of_messages')

```

```python
# Calculate the volume of internet traffic used by each user per month. Save the result.
internet_data['month'] = pd.to_datetime(internet_data['session_date']).dt.month
internet_per_month = internet_data.groupby(['user_id', 'month'])['mb_used'].sum().reset_index(name='total_mb_used')

```

[Put the aggregate data together into one DataFrame so that one record in it would represent what an unique user consumed in a given month.]

```python
# Merge the data for calls, minutes, messages, and internet based on user_id and month
user_usage = pd.merge(calls_per_month, minutes_per_month, on=['user_id', 'month'], how='outer')
user_usage = pd.merge(user_usage, messages_per_month, on=['user_id', 'month'], how='outer')
user_usage = pd.merge(user_usage, internet_per_month, on=['user_id', 'month'], how='outer')

# Replace NaN values with 0 for months with no usage
user_usage.fillna(0, inplace=True)
```

```python
# Add the plan information
user_usage = pd.merge(user_usage, users_data[['user_id', 'plan']], on='user_id', how='left')
```

[Calculate the monthly revenue from each user (subtract the free package limit from the total number of calls, text messages, and data; multiply the result by the calling plan value; add the monthly charge depending on the calling plan). N.B. This might not be as trivial as just a couple of lines given the plan conditions! So, it's okay to spend some time on it.]

```python
def calculate_monthly_revenue(row):
    plan_details = plans_data[plans_data['plan_name'] == row['plan']].iloc[0]
    
    minutes_limit = plan_details['minutes_included']
    messages_limit = plan_details['messages_included']
    data_limit = plan_details['mb_per_month_included'] / 1024  # Ensuring conversion from MB to GB

    extra_minutes = max(0, row['total_minutes'] - minutes_limit)
    extra_messages = max(0, row['number_of_messages'] - messages_limit)
    extra_data_gb = max(0, np.ceil((row['total_mb_used'] / 1024) - data_limit))  # Convert MB to GB and calculate excess
    
    extra_minutes_cost = extra_minutes * plan_details['usd_per_minute']
    extra_messages_cost = extra_messages * plan_details['usd_per_message']
    extra_data_cost = extra_data_gb * plan_details['usd_per_gb']
    
    total_revenue = (plan_details['usd_monthly_pay'] +
                     extra_minutes_cost +
                     extra_messages_cost +
                     extra_data_cost)
    
    # Debugging output
    print(f"User: {row['user_id']}, Plan: {row['plan']}, Total Revenue: {total_revenue}")
    
    return total_revenue

# Applying the function to the DataFrame with debugging to find abnormal results
user_usage['monthly_revenue'] = user_usage.apply(calculate_monthly_revenue, axis=1)



```
